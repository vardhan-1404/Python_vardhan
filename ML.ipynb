```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve, auc

# Set random seed
np.random.seed(420)

# Generate weight dataset
num_samples = 100
weight = np.sort(np.random.normal(loc=65, scale=29, size=num_samples))

# Generate obesity labels
obese = (np.random.rand(num_samples) < (np.argsort(weight) / num_samples)).astype(int)

# Plot data
plt.scatter(weight, obese)
plt.xlabel("Weight")
plt.ylabel("Obese (1=Yes, 0=No)")
plt.title("Weight vs Obesity")
plt.show()

# Logistic Regression Model
log_reg = LogisticRegression()
log_reg.fit(weight.reshape(-1, 1), obese)
log_reg_preds = log_reg.predict_proba(weight.reshape(-1, 1))[:, 1]

# Random Forest Model
rf_model = RandomForestClassifier()
rf_model.fit(weight.reshape(-1, 1), obese)
rf_preds = rf_model.predict_proba(weight.reshape(-1, 1))[:, 1]

# Compute ROC curve and AUC for Logistic Regression
fpr_log, tpr_log, _ = roc_curve(obese, log_reg_preds)
roc_auc_log = auc(fpr_log, tpr_log)

# Compute ROC curve and AUC for Random Forest
fpr_rf, tpr_rf, _ = roc_curve(obese, rf_preds)
roc_auc_rf = auc(fpr_rf, tpr_rf)

# Plot ROC Curves
plt.figure(figsize=(6, 6))
plt.plot(fpr_log, tpr_log, color='#377eb8', lw=2, label=f'Logistic Regression (AUC = {roc_auc_log:.2f})')
plt.plot(fpr_rf, tpr_rf, color='#4daf4a', lw=2, label=f'Random Forest (AUC = {roc_auc_rf:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison")
plt.legend(loc="lower right")
plt.show()
```
